```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits

# Propagation
from prysm.coordinates import make_xy_grid, cart_to_polar
from prysm.geometry import circle
from prysm.wavelengths import HeNe
from prysm.propagation import Wavefront

# from prysm.polynomials import hopkins
```

```python
# eres the jones pupil
pth = '/Users/jashcraft/Desktop/poke/ELT_Paper_1/ELT/Jones_Pupils/ELT_BareAg_551.fits'
jones = fits.open(pth)[0].data[...,0] + 1j* fits.open(pth)[0].data[...,1]
jones = jones[...,:2,:2]
```

```python
# Try program up the polarimetry stuff using prysm's coordinates
xi, eta = make_xy_grid(256, diameter=2)
dx = xi[1,0]-xi[0,0]
r, p = cart_to_polar(xi, eta)
A = circle(1, r)
plt.imshow(A)

# initial conditions
m = 3 # planes of symmetry parameter, trifold is most common
c = 4*np.pi # "constant proportional to external applied forces"
B = (2-m)/2
t = 2*B*p

ret = 2*np.pi/HeNe
# c = ret
realcomp = np.cos(c*r**(m-2)/2)
# realcomp = np.cos(ret/2)
real = np.array([[realcomp,0*realcomp],
                 [0*realcomp,realcomp]])

imagcomp = 1j*np.sin(c*r**(m-2)/2)#*(2*(2-m)*p)
# imagcomp = 1j*np.sin(ret/2)*(2*t)
imag = np.array([[imagcomp*np.cos(t),-imagcomp*np.sin(t)],
                 [-imagcomp*np.sin(t),-imagcomp*np.cos(t)]])

J = real + imag
J *= A
print(J.shape)
print('moving axis for matrix broadcasting')
J = np.moveaxis(np.moveaxis(J,2,0),3,1)
print(J.shape)
```

```python
def plotjones(j):
    
    fig,ax = plt.subplots(ncols=4,nrows=2,figsize=[10,5])
    absj = np.abs(j)
    angj = np.angle(j)
    
    print(angj.shape)
    
    for a in range(2):
        for i in range(2):
            for j in range(2):

                print(i,j)

                if a == 0:
                    cmap = ax[i,j].imshow(absj[...,i,j])
                    fig.colorbar(cmap,ax=ax[i,j])
                else:
                    ax[i,j+2].imshow(angj[...,i,j])
                    fig.colorbar(cmap,ax=ax[i,j+2])
    
    plt.show()
    
plotjones(J)
```

```python
Ein = np.array([1,1j])

RHC = np.array([[1,1j],[-1j,1]])
J = RHC @ J

# HOR = np.array([[1,0],[0,0]])
# J = HOR @ J


Eout1 = Ein[0]*J[...,0,0] + Ein[1]*J[...,0,1]
Eout2 = Ein[0]*J[...,1,0] + Ein[1]*J[...,1,1]

# Add the circular analyzer term


Eouttot = np.abs(Eout1)**2 + np.abs(Eout2)**2 
Eout_analyzed = 0

plt.figure()
plt.imshow(np.abs(Eout1))
plt.colorbar()

plt.figure()
plt.imshow(np.abs(Eout2))
plt.colorbar()

plt.figure()
plt.imshow(Eouttot,cmap='gray')
plt.colorbar()

Eout1 /= np.sum(np.abs(Eout1))

# Load the polarizations into a wavefront
wf = Wavefront.from_amp_and_phase(np.abs(Eout1), np.angle(Eout1), HeNe, dx) # wf == P
```

```python
E = wf.focus(100)
psf = E.intensity
fno = 10
psf_radius = 1.22*HeNe*fno
psf.plot2d(xlim=psf_radius*5, log=False, cmap='gray',
           clim=None, interpolation='bicubic')
```
